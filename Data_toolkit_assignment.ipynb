```python
'''3. Explain the following terms
 The difference in np.array, np.asarray and np.asanyarray
 The difference between Deep copy and shallow copy'''

>>np.array():
Purpose: Creates a new NumPy array
import numpy as np
lst = [1, 2, 3]
arr1 = np.array(lst)  # Creates a new array from the list
arr2 = np.array(arr1)  # Can create a new array even if arr1 is already a NumPy array

>>np.asarray():

Purpose: Converts the input to a NumPy array, but does not make a copy
if the input is already an array.

lst = [1, 2, 3]
arr1 = np.asarray(lst)  # Creates a new array from the list
arr2 = np.asarray(arr1)  # No new array is created; arr1 and arr2 reference the same array

>>np.asanyarray():
Purpose: Similar to np.asarray(), but it allows subclassed arrays
(such as matrix objects) to be preserved. This means it won't convert arrays
that inherit from np.ndarray to a base ndarray

matrix = np.matrix([[1, 2], [3, 4]])
arr = np.asanyarray(matrix)  # Preserves the matrix class


'''Difference Between Deep Copy and Shallow Copy'''

>>Shallow Copy:
Definition: A shallow copy of an object is a new object, but it copies only the
references of the original objectâ€™s elements. If the elements of the original
object are mutable objects (like lists or arrays), changes to those objects will
be reflected in both the original and the shallow copy.

import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = arr1.view()  # Shallow copy (view)
arr2[0] = 100  # Changes in arr2 affect arr1
print(arr1)  # Output: [100   2   3]

>>Deep Copy:
Definition: A deep copy creates a completely independent copy of the object,
including recursively copying all elements. Changes to the deep copy will not
affect the original object, and vice versa.
arr1 = np.array([1, 2, 3])
arr2 = arr1.copy()  # Deep copy
arr2[0] = 100  # Changes in arr2 do not affect arr1
print(arr1)  # Output: [1 2 3]
```

```python
'''14.Create a time-series dataset in a Pandas DataFrame with columns: 'Date',
'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
a)Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes
(left y-axis for 'Temperature' and right y-axis for 'Humidity').
b) Label the x-axis as 'Date'.
c) Set the title of the plot as 'Temperature and Humidity Over Time'.'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a time-series dataset
date_range = pd.date_range(start='2024-01-01', periods=150, freq='D')  # 100 days starting from Jan 1, 2024
temperature = np.random.uniform(low=15, high=40, size=150)  # Random temperatures between 15Â°C and 30Â°C
humidity = np.random.uniform(low=40, high=90, size=150)  # Random humidity between 40% and 90%

# Create the DataFrame
data = {
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
}
df = pd.DataFrame(data)

# a) Plot the 'Temperature' on the left y-axis
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)', color='red')
plt.plot(df['Date'], df['Temperature'], color='green', label='Temperature')
plt.tick_params(axis='y', labelcolor='red')

# b) Create a second y-axis for 'Humidity' (right y-axis)
ax2 = plt.twinx()  # Create another y-axis sharing the same x-axis
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.plot(df['Date'], df['Humidity'], color='blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='blue')


# c) Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

```python
'''15.Create a NumPy array data containing 1000 samples from a normal distribution.
Perform the following tasks using Matplotlib:
 a) Plot a histogram of the data with 30 bins.
 b) Overlay a line plot representing the normal distribution's probability density function (PDF).
 c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
 d) Set the title of the plot as 'Histogram with PDF Overlay'.'''


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

# a) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30,alpha=0.6,density='True',color='r', label='Histogram')

# b) Overlay a line plot representing the normal distribution's probability density function (PDF)
xmin, xmax = plt.xlim() # Get the limits of the x-axis
x = np.linspace(xmin, xmax, 100)  # Create an array of x values
p = norm.pdf(x, loc=0, scale=1)  # Calculate the PDF for the normal distribution
plt.plot(x, p, 'k', linewidth=2, label='PDF')  # Overlay the PDF line

# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot as 'Histogram with PDF Overlay'
plt.title('Histogram with PDF Overlay')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
'''17.Create a Seaborn scatter plot of two random arrays, color points based on their
position relative to the origin (quadrants), add a legend, label the axes, and set
the title as 'Quadrant-wise Scatter Plot'.'''

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate two random arrays
x = np.random.randn(100)
y = np.random.randn(100)
data = pd.DataFrame({'x': x, 'y': y})

# Define a function to assign quadrants
def get_quadrant(row):
    if row['x'] >= 0 and row['y'] >= 0:
        return 'Quadrant I'
    elif row['x'] < 0 and row['y'] >= 0:
        return 'Quadrant II'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Apply the function to assign quadrants to each point
data['quadrant'] = data.apply(get_quadrant, axis=1)

# Create a scatter plot using Seaborn
sns.scatterplot(data=data, x='x', y='y', hue='quadrant', palette='Set1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrants')
plt.show()
```

```python
'''18.With Bokeh, plot a line chart of a sine wave function, add grid lines,
label the axes, and set the title as 'Sine Wave Function'''

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()

# Generate x values (from 0 to 4Ï€) and their corresponding sine values
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave Function", width=700, height=400)

# Plot the sine wave (line chart)
p.line(x, y, line_width=2, color="blue", legend_label="Sine wave")

# Add grid lines (enabled by default in Bokeh)
p.grid.grid_line_color = "gray"

# Label the axes
p.xaxis.axis_label = 'X-axis (Radians)'
p.yaxis.axis_label = 'Y-axis (Amplitude)'

# Set title and customize its font size
p.title.text_font_size = '16pt'

# Display the plot
show(p)
```

```python
'''19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'''

import random
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Viridis256
output_notebook()

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [random.randint(10, 100) for _ in categories]
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add vertical bars
p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color=factor_cmap('categories', palette=Viridis256,
                                                  factors=categories))

# Add hover tooltips to display the exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Customize the plot appearance
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.xaxis.major_label_orientation = 1

# Show the plot
show(p)
```

```python
'''20.Using Plotly, create a basic line plot of a randomly generated dataset,
label the axes, and set the title as 'Simple Line Plot'''

import numpy as np
import plotly.graph_objects as go

# Generate a random dataset
x = np.arange(0, 100)  # x values (0 to 99)
y = np.random.randn(100).cumsum()  # y values: cumulative sum of random values

# Create a line plot
fig = go.Figure()

# Add a trace for the line plot
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Label the axes
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show the plot
fig.show()
```
